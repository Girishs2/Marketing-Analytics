2. Performing Customers/sellers Segmentation

a. Divide the customers into groups based on the revenue generated

```python
cust_segm = pd.cut(x=(df7.groupby('customer_unique_id')['Revenue'].sum()), bins=5,\
                   labels=['very_low_spend','low_spend','medium_spend','high_spend','very_high_spend']).reset_index()
```

```python
df7 = pd.merge(df7,cust_segm, on='customer_unique_id', how='inner')
df7.head(2)
```

```python
df7.rename(columns={'Revenue_x':'Revenue','Revenue_y':'cust_segm'}, inplace=True)
```

```python
df7.head(2)
```

```python
df7.cust_segm.value_counts()*100/df7.customer_unique_id.size
```

b. Divide the sellers into groups based on the revenue generated 

```python
seller_segm = pd.cut(x=(df7.groupby('seller_id')['Revenue'].sum()), bins=5,\
                     labels=['very_low_spend','low_spend','medium_spend','high_spend','very_high_spend']).reset_index()
```

```python
df7 = pd.merge(df7, seller_segm, on='seller_id', how='inner')
df7
```

```python
df7.rename(columns={'Revenue_x':'Revenue','Revenue_y':'seller_segm'}, inplace=True)
```

```python
df7
```
